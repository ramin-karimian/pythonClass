```python
#define df_class with dictionary and assign student Id as index
d={"Student":["a","b","c","d","e"],
    "Math_grade":[13,10,18,20,7],
    "Physics_grade":[2,12,20,15,10]}
studentId=[1001,1002,1003,1004,1005]
```

```python
df_class=pd.DataFrame(d,index=studentId)
df_class
```

```python
#sum()
df_class.sum()
```

```python
#sum axis 1
df_class.sum(axis=1)
```

```python
# min and min axis =1
df_class.min(axis=0)
df_class.min(axis=1)
```

```python
#idxmin axis 0 and 1
df_class[["Math_grade","Physics_grade"]].idxmin(axis=1)
```

```python
#describe()
df_class.describe()
df_class.apply(pd.DataFrame.describe,axis=1)
```

```python
df_class.mean(axis=1)
```
